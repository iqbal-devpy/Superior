```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load Image
img = cv2.imread('image.png') 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  

# Display the image
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
resized = cv2.resize(img, (300, 300))
```

```python
cropped = img[50:200, 50:200]  
```

```python
rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
```

```python
flipped = cv2.flip(img, 1)  

# Display all modifications
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].imshow(resized)
axs[0, 0].set_title("Resized Image")
axs[0, 1].imshow(cropped)
axs[0, 1].set_title("Cropped Image")
axs[1, 0].imshow(rotated)
axs[1, 0].set_title("Rotated Image")
axs[1, 1].imshow(flipped)
axs[1, 1].set_title("Flipped Image")
plt.show()
```

```python
draw_img = img.copy()
cv2.line(draw_img, (50, 50), (250, 50), (255, 0, 0), 5)
cv2.rectangle(draw_img, (50, 50), (250, 250), (0, 255, 0), 3)
cv2.circle(draw_img, (150, 150), 50, (0, 0, 255), -1)
cv2.putText(draw_img, "OpenCV", (50, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

plt.imshow(draw_img)
plt.title("Drawing on Image")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
edges = cv2.Canny(gray, 100, 200)

# Display results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title("Grayscale Image")
axs[1].imshow(binary, cmap='gray')
axs[1].set_title("Binary Thresholding")
axs[2].imshow(edges, cmap='gray')
axs[2].set_title("Edge Detection (Canny)")
plt.show()
```

```python
contours, _ = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

plt.imshow(contour_img)
plt.title("Contours Detection")
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray, 1.1, 4)
face_img = img.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 0, 0), 2)

plt.imshow(face_img)
plt.title("Face Detection")
plt.show()
```
